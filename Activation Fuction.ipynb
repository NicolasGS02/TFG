# ActivaciГіn

## Crearemos una pequeГ±a red neuronal multivariable para ver las salidas con diferentes casos

```python
import numpy as np
import keras
from keras import ops
from keras import layers
#Trabajaremos con el dataset de iris para comprobar los diferentes resultados.
mnistDataset = keras.datasets.mnist
```

```python
(x_train, y_train), (x_test, y_test) = mnistDataset.load_data()
```

```python
x_train = x_train / 255.0
x_test = x_test / 255.0

x_train = 2.0 * x_train - 1.0
x_test  = 2.0 * x_test  - 1.0


x_train = x_train.reshape(60000, 784)
x_test = x_test.reshape(10000, 784)
```

### Creamos nuestro modelo

```python
#Modelo con activaciГіn ReLU

inputs1 = keras.Input(shape=(784,))
first_Layer1 = layers.Dense(30, activation="relu")(inputs1)
second_Layer1 = layers.Dense(30, activation="relu")(first_Layer1)
third_Layer1 = layers.Dense(30, activation="relu")(second_Layer1)
outputs1 = layers.Dense(10, activation="softmax")(third_Layer1)

RELUtestModel = keras.Model(inputs=inputs1, outputs=outputs1)

#modelo con activaciГіn tanh 
inputs2 = keras.Input(shape=(784,))
first_Layer2 = layers.Dense(30, activation="tanh")(inputs2)
second_Layer2 = layers.Dense(30, activation="tanh")(first_Layer2)
third_Layer2 = layers.Dense(30, activation="tanh")(second_Layer2)
outputs2 = layers.Dense(10, activation="softmax")(third_Layer2)
TANHtestModel = keras.Model(inputs=inputs2, outputs=outputs2)

#modelo combinando activaciГіn ReLU y tanh
inputs3 = keras.Input(shape=(784,))
first_Layer3 = layers.Dense(30, activation="tanh")(inputs3)
second_Layer3 = layers.Dense(30, activation="relu")(first_Layer3)
third_Layer3 = layers.Dense(30, activation="swish")(second_Layer3)
outputs3 = layers.Dense(10, activation="softmax")(third_Layer3)
MIXEDtestModel = keras.Model(inputs=inputs3, outputs=outputs3)
```

```python
RELUtestModel.compile(
    loss=keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    optimizer=keras.optimizers.RMSprop(),
    metrics=["accuracy"],
)
TANHtestModel.compile(
    loss=keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    optimizer=keras.optimizers.RMSprop(),
    metrics=["accuracy"],
)
MIXEDtestModel.compile(
    loss=keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    optimizer=keras.optimizers.RMSprop(),
    metrics=["accuracy"],
)
```

```python
#Entrenaos modelo
print("Entrenando modelo con activaciГіn ReLU")
RELUtestModel.fit(x_train, y_train, epochs=5, batch_size=128)
print("Entrenando modelo con activaciГіn tanh")
TANHtestModel.fit(x_train, y_train, epochs=5, batch_size=128)
print("Entrenando modelo con activaciГіn mixta")
MIXEDtestModel.fit(x_train, y_train, epochs=5, batch_size=128)
```

```python
RELUtestModel.evaluate(x_test, y_test, batch_size=128)
TANHtestModel.evaluate(x_test, y_test, batch_size=128)
MIXEDtestModel.evaluate(x_test, y_test, batch_size=128)

print("Acurracy del modelo con activaciГіn ReLU: ", RELUtestModel.evaluate(x_test, y_test, batch_size=128)[1])
print("Acurracy del modelo con activaciГіn tanh: ", TANHtestModel.evaluate(x_test, y_test, batch_size=128)[1])
print("Acurracy del modelo con activaciГіn mixta: ", MIXEDtestModel.evaluate(x_test, y_test, batch_size=128)[1])
```
